```python
# Đường dẫn đến thư mục chứa ảnh gốc
ORIGINAL_DUANON_PATH = r"/content/drive/MyDrive/DPL302m_Project/data/dừa non"
```

```python
# Số lượng ảnh bạn muốn tạo
NUM_AUGMENTED_IMAGES = 600
```

```python
import os
import random
from torchvision import transforms
from PIL import Image
```

```python
# Biến đổi dữ liệu
DATA_AUGMENTATION = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
])
```

```python
# Lấy số cuối cùng trong tên ảnh gốc
last_image_number = max(
    (int(file.split('.')[0]) for file in os.listdir(ORIGINAL_DUANON_PATH) if file.endswith('.jpg')),
    default=0
)
```

```python
image_count = last_image_number + 1
original_images = [file for file in os.listdir(ORIGINAL_DUANON_PATH) if file.endswith('.jpg')]

# Lặp để tạo thêm ảnh mới
for i in range(NUM_AUGMENTED_IMAGES):
    original_image_path = os.path.join(ORIGINAL_DUANON_PATH, random.choice(original_images))
    new_image_name = f'{image_count}.jpg'
    new_image_path = os.path.join(ORIGINAL_DUANON_PATH, new_image_name)
    image = Image.open(original_image_path)
    augmented_image = DATA_AUGMENTATION(image)
    augmented_image = augmented_image.convert('RGB')
    augmented_image.save(new_image_path, 'JPEG')
    image_count += 1

print("Finished")
```

```python
import os
import shutil
import random
```

```python
# Đường dẫn đến thư mục chứa ảnh gốc
DATA_DIR = r'/content/drive/MyDrive/DPL302m_Project/data/other'
```

```python
TRAIN_DIR = r'/content/drive/MyDrive/DPL302m_Project/train/loại khác'
VAL_DIR = r'/content/drive/MyDrive/DPL302m_Project/val/loại khác'
TEST_DIR = r'/content/drive/MyDrive/DPL302m_Project/test/loại khác'
```

```python
# Tạo thư mục nếu chưa tồn tại
os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(VAL_DIR, exist_ok=True)
os.makedirs(TEST_DIR, exist_ok=True)
```

```python
# Lấy danh sách tất cả các tệp hình ảnh trong thư mục dữ liệu
IMAGE_FILES = os.listdir(DATA_DIR)
```

```python
# Tính số lượng tệp cho mỗi phần
TOTAL_IMAGES = len(IMAGE_FILES)
TRAIN_SPLIT = int(0.8 * TOTAL_IMAGES)
VAL_SPLIT = int(0.1 * TOTAL_IMAGES)
```

```python
# Tính số lượng tệp cho mỗi phần
TOTAL_IMAGES = len(IMAGE_FILES)
TRAIN_SPLIT = int(0.8 * TOTAL_IMAGES)
VAL_SPLIT = int(0.1 * TOTAL_IMAGES)
```

```python
# Chia tệp
TRAIN_FILES = IMAGE_FILES[:TRAIN_SPLIT]
VAL_FILES = IMAGE_FILES[TRAIN_SPLIT:TRAIN_SPLIT + VAL_SPLIT]
TEST_FILES = IMAGE_FILES[TRAIN_SPLIT + VAL_SPLIT:]
```

```python
# Di chuyển vào mục tương ứng
for file in TRAIN_FILES:
    src = os.path.join(DATA_DIR, file)
    dst = os.path.join(TRAIN_DIR, file)
    shutil.copy(src, dst)

for file in VAL_FILES:
    src = os.path.join(DATA_DIR, file)
    dst = os.path.join(VAL_DIR, file)
    shutil.copy(src, dst)

for file in TEST_FILES:
    src = os.path.join(DATA_DIR, file)
    dst = os.path.join(TEST_DIR, file)
    shutil.copy(src, dst)

print("Finished")
```
